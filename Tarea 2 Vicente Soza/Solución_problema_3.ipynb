```python
# Detector de Puntaje gamer

num1 = int(input("Ingrese el primer numero: "))
num2 = int(input("Ingrese el segundo numero: "))

suma = num1 + num2

if suma < 50:
  print("Resultado: Nivel Principiante")
elif suma < 80:
  print("Resultado: Nivel Intermedio")
else:
  print("Resultado: Nivel Pro")
```
